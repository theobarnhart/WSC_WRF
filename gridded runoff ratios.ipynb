```python
wy = 2006

P = np.load('./data/precip_wy%s.npz'%wy)['arr_0'] #*0.001 # mm to m
bf = np.load('./data/groundwater_wy%s.npz'%wy)['arr_0'] #*0.001 # mm to m
ru = np.load('./data/runoff_wy%s.npz'%wy)['arr_0'] #*0.001 # mm to m

Q = bf[0]+ru[0]
```

```python
RR = Q/P[0]
```

```python
plt.figure(figsize=(6,7))
plt.pcolormesh(RR,cmap='jet_r')
plt.axis('equal')
cb = plt.colorbar()
cb.set_label('Runoff Ratio', fontsize = 15)
plt.xlabel('East-West',fontsize=14)
plt.ylabel('North-South',fontsize=14)
plt.title('Water Year %s'%wy,fontsize=16,fontweight='bold')
```

```python
plt.figure(figsize=(6,7))
plt.pcolormesh(RR,vmin=0,vmax=1,cmap='jet_r')
plt.axis('equal')
cb = plt.colorbar()
cb.set_label('Runoff Ratio, limited to 0-1.0', fontsize = 15)
plt.xlabel('East-West',fontsize=14)
plt.ylabel('North-South',fontsize=14)
plt.title('Water Year %s'%wy,fontsize=16,fontweight='bold')
```

```python
plt.figure(figsize=(6,7))
plt.pcolormesh(P[0],cmap='jet_r',vmin = 0,vmax=P.max())
plt.axis('equal')
cb = plt.colorbar()
cb.set_label('Water Year Precip. [mm]', fontsize = 15)
plt.xlabel('East-West',fontsize=14)
plt.ylabel('North-South',fontsize=14)
plt.title('Water Year %s'%wy,fontsize=16,fontweight='bold')
```

```python
plt.figure(figsize=(6,7))
plt.pcolormesh(Q,cmap='jet_r',vmin = 0,vmax=P.max())
plt.axis('equal')
cb = plt.colorbar()
cb.set_label('Water Year UDROFF + SFROFF [mm]', fontsize = 15)
plt.xlabel('East-West',fontsize=14)
plt.ylabel('North-South',fontsize=14)
plt.title('Water Year %s'%wy,fontsize=16,fontweight='bold')
```
